##### Have you noticed that the type of Year is int64. But pandas has a different type to work with Time Series. Let's see it now.

### Step 5. Convert the type of the column Year to datetime64

```python
crime['Year'] = pd.to_datetime(crime["Year"], format='%Y')
```

### Step 6. Set the Year column as the index of the dataframe

```python
crime.set_index('Year', inplace=True)
```

```python
crime.head()
```

### Step 7. Delete the Total column

```python
crime.drop('Total',axis=1,inplace=True)
```

```python
crime.head()
```

### Step 8. Group the year by decades and sum the values

#### Pay attention to the Population column number, summing this column is a mistake

```python
crime_reset = crime.reset_index()
crime_reset['year'] = crime_reset['Year'].dt.year
crime_reset['decade'] = crime_reset['year'].apply(lambda x: (x//10)*10)
crime_reset.head()
```

```python
# OR

# crime_reset = crime.reset_index()
# crime_reset['year'] = crime_reset['Year'].dt.year
# crime_reset['decade'] = (crime_reset['year'] // 10) * 10

# crime_reset.head()
```

```python
grouped_crime = crime_reset.groupby('decade')[['Violent', 'Property', 'Murder', 'Forcible_Rape', 'Robbery', 'Aggravated_assault', 'Burglary', 'Larceny_Theft', 'Vehicle_Theft']].sum()
grouped_crime
```

### Step 9. What is the most dangerous decade to live in the US?

```python
grouped_crime['total'] = grouped_crime[['Violent', 'Property', 'Murder', 'Forcible_Rape', 'Robbery', 'Aggravated_assault', 'Burglary', 'Larceny_Theft', 'Vehicle_Theft']].sum(axis=1)
grouped_crime.head()
```

```python
grouped_crime['Total'] = grouped_crime[['Violent', 'Property', 'Murder', 'Forcible_Rape', 'Robbery', 'Aggravated_assault', 'Burglary', 'Larceny_Theft', 'Vehicle_Theft']].sum(axis=1)

most_dangerous_decade = grouped_crime['Total'].idxmax()

print(f"The most dangerous decade to live in the US was the {most_dangerous_decade}s")
```
