```python
## Z score computation

outliers = []

def detect_outliers(data):
  threshold = 3 ##3rd standard deviation
  mean = np.mean(data)
  print(mean)
  std = np.std(data)
  print(std)

  for i in data:
    z_score = (i-mean)/std
    if np.abs(z_score)> threshold:
      outliers.append(i)

  return outliers
```

```python
detect_outliers(dataset)
```

#IQR (Interquartile Range)
#Find outliers using IQR

1. Sort the data
2. Calculate Q1 and Q3
3. IQR (Q3-Q1)
4. Find the lower fence (Q1 - 1.5(IQR))
5. Find the upper fence (Q3 + 1.5(IQR))


```python
q1, q3 = np.percentile(dataset, [25,75])
```

```python
print(q1,q3)
```

```python
IQR = q3-q1
print(IQR)
```

```python
## Find the lower and higher fence
lower_fence = q1 - (1.5*IQR)
higher_fence = q3 + (1.5*IQR)
```

```python
print(lower_fence, higher_fence)
```

```python
for i in dataset:
  if i<lower_fence or i>higher_fence:
    print(i)
```

```python
# Set random seed for reproducibility
np.random.seed(42)
```

```python
# 1. Create a population: 1000 students with test scores normally distributed
population = np.random.normal(loc=70, scale=10, size=1000)  # mean=70, std=10
```

```python
# 2. Take a random sample of 100 students
sample = np.random.choice(population, size=100, replace=False)
```

```python
# 3. Compare population mean and sample mean
population_mean = np.mean(population)
sample_mean = np.mean(sample)
```

```python
print(f"Population Mean: {population_mean:.2f}")
print(f"Sample Mean: {sample_mean:.2f}")
```

```python
# 4. Plot the distribution
plt.figure(figsize=(10, 5))
sns.histplot(population, bins=30, color='skyblue', label='Population', kde=True)
sns.histplot(sample, bins=30, color='orange', label='Sample', kde=True)
plt.axvline(population_mean, color='blue', linestyle='--', label='Population Mean')
plt.axvline(sample_mean, color='orange', linestyle='--', label='Sample Mean')
plt.title('Population vs Sample Distribution')
plt.legend()
plt.show()
```

Dimensions in Arrays
A dimension in arrays is one level of array depth (nested arrays).

nested array: are arrays that have arrays as their elements.

Types:
0-D Arrays
1-D Arrays
2-D Arrays
3-D arrays

NumPy has a whole sub module dedicated towards matrix operations called numpy.mat

```python
import numpy as np

a = np.array(42)
b = np.array([1, 2, 3, 4, 5])
c = np.array([[1, 2, 3], [4, 5, 6]])
d = np.array([[[1, 2, 3], [4, 5, 6]], [[1, 2, 3], [4, 5, 6]]])

print(a.ndim)
print(b.ndim)
print(c.ndim)
print(d.ndim)
```

Higher Dimensional Arrays
An array can have any number of dimensions.

When the array is created, you can define the number of dimensions by using the ndmin argument.

```python
import numpy as np

arr = np.array([1, 2, 3, 4], ndmin=5)

print(arr)
print('number of dimensions :', arr.ndim)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Create a right-skewed dataset using exponential distribution
np.random.seed(0)
data = np.random.exponential(scale=2.0, size=1000)

# Calculate Mean, Median, and Mode
mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data, keepdims=True).mode[0]

# Plot histogram
plt.figure(figsize=(12, 6))
plt.hist(data, bins=50, color='lightgray', edgecolor='black', alpha=0.7)
plt.axvline(mean, color='red', linestyle='--', linewidth=2, label=f"Mean: {mean:.2f}")
plt.axvline(median, color='blue', linestyle='--', linewidth=2, label=f"Median: {median:.2f}")
plt.axvline(mode, color='green', linestyle='--', linewidth=2, label=f"Mode: {mode:.2f}")

plt.title("Mean vs Median vs Mode on Skewed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()
```
